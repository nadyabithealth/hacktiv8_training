## Area Pots (Series/Dataframe) <a id="12"></a>

**What is a line plot and why use it?**

An Area chart or area plot is a type of plot which displays information as a series of data points called 'markers' connected by straight line segments. It is a basic type of chart common in many fields. Use line plot when you have a continuous data set. These are best suited for trend-based visualizations of data over a period of time.

**Questions:**

How does the distribution of major crime categories change over the years (2008 to 2016) in London?

```python
df.year.value_counts()
```

```python
# Write your function below

# Graded-Funtion Begin (~1 Lines)
# Filter data for 2008 to 2016
crime_years = df[(df['year'] >= 2008) & (df['year'] <= 2016)]

# Group by year and major_category, then sum the values
grouped_data = crime_years.groupby(['year', 'major_category'])['value'].sum().unstack().fillna(0)

# sorted the column
sorted_columns = grouped_data.sum().sort_values(ascending=True).index
grouped_data = grouped_data[sorted_columns]

# Custom color
palette = sns.color_palette("Set3")

# Plot area chart
plt.figure(figsize=(12, 8))
grouped_data.plot(kind='area', stacked=True, alpha=0.8, color=palette)
plt.legend(title='Major Crime Categories', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Distribution of Major Crime Categories Over the Years (2008-2016)')
plt.xlabel('Year')
plt.ylabel('Total Incidents')

# Graded-Funtion End
plt.show()
```

```python
grouped_data
```

**Insight:**

1. Pada umumnya, total kejahatan menunjukkan tren penurunan dari tahun 2008 higga 2016 khususnya untuk beberapa kejahatan seperti Sexual Offences, Fraud or Forgery, Other Notifiable Offences, Robbery, Drugs, Criminal Damage, Burglary.
2. Kategori kejahatan seperti Robbery dan Theft and handling cendrung menunjukkan penurunan yang konsisten.
3. Beberapa kategori kejahatan, seperti Violence Against the Person dan Theft and Handling, menunjukkan tren peningkatan seiring waktu, terutama pada tahun 2011 dan seterusnya.

## Histogram

A histogram is a way of representing the frequency distribution of numeric dataset. The way it works is it partitions the x-axis into bins, assigns each data point in our dataset to a bin, and then counts the number of data points that have been assigned to each bin. So the y-axis is the frequency or the number of data points in each bin. Note that we can change the bin size and usually one needs to tweak it so that the distribution is displayed nicely.

**Question:**

What is the distribution of the total sum of major crime values across different months?

```python
# Write your function below

# Graded-Funtion Begin (~2 Lines)
# Group by month and sum the major crime values in 2016
df_2016 = df[df['year'] == 2016]
crimes_monthly_sum = df_2016.groupby('month')['value'].sum()

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(crimes_monthly_sum.index, bins=range(1, 14), weights=crimes_monthly_sum, color='lightblue', edgecolor='grey', align='left')
# add value label on top of histogram
for i, value in enumerate(crimes_monthly_sum):
    plt.text(i + 1, value + 1000, f'{value/1000:.0f}k', ha='center', va='bottom', fontsize=10, color='black')
plt.title('Monthly Sum of Major Crime Values (Year 2016)')
plt.xlabel('Month')
plt.ylabel('Sum of Crime Values')
plt.xticks(range(1, 13))

# Graded-Funtion End
plt.show()
```

```python
crimes_monthly_sum
```

**Insight:**

1. Terjadi fluktuasi bulanan dengan puncak tertinggi pada bulan Juli (65,519 kejadian) dan puncak terendah pada bulan Februari (56,697 kejadian).
2. Grafik menunjukkan peningkatan kejahatan dari Mei hingga Juli, mencapai puncak tertinggi pada bulan Juli.
3. Setelah puncak tertinggi pada bulan Juli, terjadi penurunan jumlah kejahatan pada bulan-bulan berikutnya.

## Bar Charts (Dataframe) <a id="10"></a>

A bar plot is a way of representing data where the *length* of the bars represents the magnitude/size of the feature/variable. Bar graphs usually represent numerical and categorical variables grouped in intervals. 

To create a bar plot, we can pass one of two arguments via `kind` parameter in `plot()`:

* `kind=bar` creates a *vertical* bar plot
* `kind=barh` creates a *horizontal* bar plot

**Question:**

What are the top ten boroughs with the highest major crime values over all the years?

```python
# Write your function below

# Graded-Funtion Begin (~1 Lines)
# Group by borough and major_category, then sum the values
grouped_data = df.groupby(['borough', 'major_category'])['value'].sum().unstack()

# Sort in descending order and select the top ten
top_ten_boroughs = grouped_data.sum(axis=1).sort_values(ascending=False).head(10).index
grouped_data_top_ten = grouped_data.loc[top_ten_boroughs].iloc[::-1]

# Plot
plt.figure(figsize=(12, 8))
ax = grouped_data_top_ten.plot(kind='barh', stacked=True, colormap='Set3', edgecolor='white')

# Add value labels right beside the bars
for index, value in enumerate(grouped_data_top_ten.sum(axis=1)):
    plt.text(value, index, f'{value/1000:.1f}k', ha='left', va='center', fontsize=10, color='black')

plt.legend(title='Major Crime Categories', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Top Ten Boroughs of Major Crime Values (All Years)')
plt.xlabel('Sum of Crime Values')
plt.ylabel('Borough')

# Graded-Funtion End
plt.show()
```

```python
grouped_data_top_ten
```

**Insight:**

1. Westminster menonjol sebagai borough dengan tingkat kejahatan tertinggi, terutama dalam Theft and Handling serta Violence Against the Person.
2. Kategori kejahatan Theft and Handling dan Violence againt Person mendominasi di sebagian besar borough, mencerminkan pola kejahatan perkotaan.
3. Sexual Offences dan Fraud or Forgery cenderung memiliki tingkat kejahatan yang rendah di seluruh borough.

## Pie Charts <a id="6"></a>

A `pie chart` is a circualr graphic that displays numeric proportions by dividing a circle (or pie) into proportional slices. You are most likely already familiar with pie charts as it is widely used in business and media. We can create pie charts in Matplotlib by passing in the `kind=pie` keyword.

**Question:**

How the Proportion of Total Crimes by Major Category?

```python
# Calculate total crimes per major category
total_crimes_by_major_category = df.groupby('major_category')['value'].sum()

# Set a small percentage threshold for grouping
threshold = 5.0  # Categories with less than 5% will be grouped into "Others"

# Create a new dataframe with "Others" category
small_categories = total_crimes_by_major_category[total_crimes_by_major_category / total_crimes_by_major_category.sum() * 100 < threshold]
df['major_category_new'] = df['major_category'].replace(small_categories.index, 'Others')

# Recalculate total crimes per major category
total_crimes_by_major_category = df.groupby('major_category_new')['value'].sum()

# Create a pie chart showing the top 5 categories
top_categories = total_crimes_by_major_category.sort_values(ascending=False).head(5)
top_categories['Others'] = total_crimes_by_major_category[~total_crimes_by_major_category.index.isin(top_categories.index)].sum()

# Create a pie chart
plt.figure(figsize=(10, 8))
plt.pie(top_categories, labels=top_categories.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
plt.title('Proportion of Total Crimes by Major Category')
plt.show()
```

```python
total_crimes_by_major_category
```

**Insight:**

Data kejahatan mencerminkan dominasi "Theft and Handling" dengan jumlah 2.66 juta kejadian, diikuti oleh "Burglary" dan "Drugs" yang mencapai 754,293 dan 470,765 kejadian. "Violence Against the Person" juga signifikan dengan 1,558,081 kejadian. Kategori "Others" sebagai kejahatan sisanya.

## Box Plots <a id="8"></a>

A `box plot` is a way of statistically representing the *distribution* of the data through five main dimensions: 

- **Minimun:** Smallest number in the dataset.
- **First quartile:** Middle number between the `minimum` and the `median`.
- **Second quartile (Median):** Middle number of the (sorted) dataset.
- **Third quartile:** Middle number between `median` and `maximum`.
- **Maximum:** Highest number in the dataset.

**Question:**

How Distribution of Number of Crimes for "Violence Against the Person" in Top 5 Boroughs?

```python
# Filter the DataFrame for the major category 'Violence Against the Person'
df_theft = df[df['major_category'] == 'Violence Against the Person']

# Get the top 5 boroughs based on the sum of values
top_5_boroughs = df_theft.groupby('borough')['value'].sum().nlargest(5).index
df_theft_top5 = df_theft[df_theft['borough'].isin(top_5_boroughs)]

plt.figure(figsize=(12, 8))

# Create a boxplot with Set3 colormap
colors = plt.cm.Set3.colors
boxplot = plt.boxplot([df_theft_top5[df_theft_top5['borough'] == borough]['value'] for borough in top_5_boroughs],
                      labels=top_5_boroughs, patch_artist=True)

# Assign colors to each box
for patch, color in zip(boxplot['boxes'], colors):
    patch.set_facecolor(color)

# Set labels and title
plt.xlabel('Borough')
plt.ylabel('Number of Crimes')
plt.title('Distribution of Number of Crimes for "Violence Against the Person" in Top 5 Boroughs')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

plt.show()
```

**Insight:**

Berdasarkan grafik boxplot di atas tentang distribusi kejahatan "Violence Against the Person" pada top 5 borough dengan jumlah total kejahatan terbanyak, dapat disimpulkan bahwa:

* Titik minimum (batas bawah) hampir seragam di seluruh borough, menunjukkan adanya kecenderungan serupa dalam distribusi nilai terendah kejahatan.
* Q1, Q2 (median), dan Q3 juga menunjukkan keseragaman distribusi kejahatan di antara kelima borough tersebut.
* Terdapat variasi yang cukup signifikan dalam jumlah outlier di setiap borough. Westminster memiliki jumlah outlier terbanyak, menunjukkan adanya perbedaan yang mencolok dalam tingkat kejahatan di borough tersebut.

## Scatter Plots <a id="10"></a>

A `scatter plot` (2D) is a useful method of comparing variables against each other. `Scatter` plots look similar to `line plots` in that they both map independent and dependent variables on a 2D graph. While the datapoints are connected together by a line in a line plot, they are not connected in a scatter plot. The data in a scatter plot is considered to express a trend. With further analysis using tools like regression, we can mathematically calculate this relationship and use it to predict trends outside the dataset.

**Question:**

How the Number of Crimes Each Month in 2011?

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter data for the year 2011
data_2016 = df[df['year'] == 2016]

# Group by month and major category, calculate total crimes
monthly_crimes_2016 = data_2016.groupby(['month', 'major_category'])['value'].sum().reset_index()

# Create scatter plot with hue based on major category
plt.figure(figsize=(12, 8))
sns.scatterplot(data=monthly_crimes_2016, x='month', y='value', hue='major_category', palette='Set3', s=90)
plt.legend(title='Major Crime Categories', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Number of Crimes Each Month in 2016')
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.show()
```

**Insights**

1. Meskipun jumlah kejahatan "Violence Against the Person" lebih sedikit, pola fluktuasinya mirip dengan "Theft and Handling." Korelasi positif menunjukkan bahwa fluktuasi pada keduanya memiliki hubungan positif, sehingga perubahan pada satu kategori mencerminkan perubahan serupa pada kategori lainnya.

## Folium

Folium is a powerful Python library that helps you create several types of Leaflet maps. The fact that the Folium results are interactive makes this library very useful for dashboard building.

From the official Folium documentation page:

> Folium builds on the data wrangling strengths of the Python ecosystem and the mapping strengths of the Leaflet.js library. Manipulate your data in Python, then visualize it in on a Leaflet map via Folium.

> Folium makes it easy to visualize data that's been manipulated in Python on an interactive Leaflet map. It enables both the binding of data to a map for choropleth visualizations as well as passing Vincent/Vega visualizations as markers on the map.

> The library has a number of built-in tilesets from OpenStreetMap, Mapbox, and Stamen, and supports custom tilesets with Mapbox or Cloudmade API keys. Folium supports both GeoJSON and TopoJSON overlays, as well as the binding of data to those overlays to create choropleth maps with color-brewer color schemes.

```python
#!conda install -c conda-forge folium=0.5.0 --yes
#!pip install folium
import folium

print('Folium installed and imported!')
```

```python
import folium
from folium import Choropleth
import json

# Assuming df_crimes_val is your DataFrame
borough_crimes = df.groupby('borough')['value'].sum().reset_index()
boroughs_in_df = borough_crimes['borough'].unique()

# Load GeoJSON data (can be downloaded from: https://github.com/radoi90/housequest-data/tree/master)
geojson_path = 'london_boroughs.geojson'
with open(geojson_path, 'r') as geojson_file:
    geojson_data = json.load(geojson_file)

# Filter GeoJSON data to include only the relevant boroughs
filtered_geojson = {
    "type": geojson_data["type"],
    "features": [feature for feature in geojson_data["features"] if feature["properties"]["name"] in boroughs_in_df]
}

# Create a Folium map centered around London
london_map = folium.Map(location=[51.509865, -0.118092], zoom_start=10)

# Create a choropleth map layer
Choropleth(
    geo_data=filtered_geojson,
    data=borough_crimes,
    columns=['borough', 'value'],
    key_on='feature.properties.name',
    fill_color='YlGnBu', #YlGnBu, YlOrRd
    fill_opacity=0.75,
    line_opacity=0.2,
    legend_name='Number of Crimes',
).add_to(london_map)

# Add borough labels
for feature in filtered_geojson['features']:
    name = feature['properties']['name']
    coords = feature['geometry']['coordinates'][0][0]  # Use the coordinates of one of the borough's points
    # Create a marker with a tooltip
    marker = folium.Marker(
        location=[coords[1], coords[0]],
        tooltip=name
    )
    marker.add_to(london_map)
    #folium.Marker(location=[coords[1], coords[0]], popup=name).add_to(london_map) # pop up with click

# Display the map in the notebook
london_map
```
